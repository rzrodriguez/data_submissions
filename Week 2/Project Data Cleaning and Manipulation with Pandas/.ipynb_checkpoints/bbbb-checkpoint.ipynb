```python
#importando bibliotecas necessarias
import os
import pandas as pd
import re
import pandas_profiling as pp
```

```python
#importando dados utilizando o encoding latin1
data = pd.read_csv('attacks.csv',encoding = 'latin1')
```

```python
#visualizando os dados
data.head()
```

```python
#visualizando os dados. É possível observar pelos dados finais que existem linhas com 'NaN'
data.tail(10)
```

```python
#com uma visao superficial já é possível ver que algumas linhas e colunas estão com valores faltando
#utilizando o comando 'isnull' e somando esta condição para checar quantos missing values cada coluna tem
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

```python
#Podemos observar que a grande maioria dos dados apresenta valores nulos. A coluna com menos dados nulos, apresenta 17021 linhas
#nulas, o que mostra que grande parte da base de 25722 linhas apresenta dados nulos. 
```

```python
#Também observa-se que algumas colunas apresentam quase que em sua totalidade dados nulos e, portanto, não acrescentam valor 
#para a análise. Para eliminar-los, utilizarei o comando 'drop'. A princípio vou eliminar as colunas Unnamed 22 e 23, que estao vazias.
# Colunas com informações desnecessarias para a formação de insights também serão removidas. São elas:
# pdf, href formula, Case Number.1, Case Number.2, original order
drop_cols_1 = ['pdf', 'href formula', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23'] 
data = data.drop(drop_cols_1, axis=1)
```

```python
data.head()
```

```python
data.columns
```

```python
#eliminando todas as linhas que contem NaN em todas as colunas
data.dropna(axis=0,how='all',inplace=True)
```

```python
data.head()
```

```python
#Padronizando os nomes das colunas
cols = list(data.columns)
print(cols)
```

```python
cols_fix = [col.strip() for col in cols]
print(cols_fix)
```
