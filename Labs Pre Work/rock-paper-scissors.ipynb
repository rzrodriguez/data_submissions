```python
#importando a biblioteca random
import random
```

```python
#variaveis
game = ["stone", "paper", "scissors"]
number_games = int(input("Maximum number of games: "))
wins_needed = int(input("Number of games a player must win to win: "))
while wins_needed > number_games:
    print("Value must be lower than maximum games.")
    wins_needed = int(input("Number of games a player must win to win: "))
machine = ""
human = ""
wins_machine = 0
wins_human = 0 
ties = 0

#random entre opcoes do jogo
def rand(r):
    return random.choice(r)

#escolha do jogador humano. So permite jogar stone, paper ou scissors
def choice():
    human = input("What is your choice? ")
    while human != "stone" and human != "paper" and human != "scissors":
        human = input("Invalid choice. What is your choice? ")
    return human

#faz o duelo e encontra o ganhador da rodada
def combat(m, h):
    if (m == "stone" and h == "stone") or (m == "paper" and h == "paper") or (m == "scissors" and h == "scissors"):
        return 0
    elif (m == "stone" and h == "scissors") or (m == "paper" and h == "stone") or (m == "scissors" and h == "paper"):
        return 1
    else:
        return 2

#mostra a jogada dos jogadores
def show():
    return print("Machine played", machine), print("Human played", human)

#mostra o placar atual
def score(wm,wh,t):
    return print("Machine: ", wm, "Human: ", wh, "Ties: ", t)


#loop do jogo
for i in range(number_games):
    if wins_machine or wins_human < wins_needed:
        machine = rand(game)
        human = choice()
        combat(machine,human)
        if combat(machine,human) == 0:
            ties += 1
        elif combat(machine,human) == 1:
            wins_machine += 1
        elif combat(machine,human) == 2:
            wins_human += 1
        show()
        score(wins_machine,wins_human, ties)
    else:
        score(wins_machine,wins_human, ties)

print("Game is over!")
```
