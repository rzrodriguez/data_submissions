# Projeto 3 - Data Pipeline - Brazilian E-Commerce Public Dataset by Olist
https://www.kaggle.com/olistbr/brazilian-ecommerce


- Dataset público de ecommerce no Brasil da Olist Store;

- Informações de 100k transações nos anos de 2016 a 2018;

- Dataset real, com informações anônimas;

- Informações de tipo do produto, preço, frete e formas de pagamento foram utilizadas.

- Quais os produtos mais demandados? 

- O ticket médio varia entre regiões do país? 

- Quais as formas de pagamentos mais utilizadas? Suas características mudam entre estados?
