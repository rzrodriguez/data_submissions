```python
#importando bibliotecas necessarias
import os
import pandas as pd
import re
import pandas_profiling as pp
```

```python
#importando dados utilizando o encoding latin1
data = pd.read_csv('attacks.csv',encoding = 'latin1')
```

```python
#visualizando os dados
data.head()
```

```python
#visualizando os dados. É possível observar pelos dados finais que existem linhas com 'NaN'
data.tail()
```

```python
#Gerando profile report pela biblioteca Pandas Profile, para ter uma visão melhor das caracteristicas dos dados
pp.ProfileReport(data)
```

```python
#com uma visao superficial já é possível ver que algumas linhas e colunas estão com valores faltando
#utilizando o comando 'isnull' e somando esta condição para checar quantos missing values cada coluna tem
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

```python
data.rename(index=str,columns = {'Sex ': 'Sex','Species ': 'Species'},inplace=True)
```

```python
#Analisando o profile report, grande parte da base de dados possui valores nulos, a princípio removerei as colunas 'Unnamed', que além de estarem vazias não fazem sentido para a análise. Também já estou removendo algumas colunas desnecessárias para formação de insights. Para as demais colunas, antes de tomar a decisão de remove-las, vou analisar as linhas que possuem poucos ou nenhum dado.
drop_cols_1 = ['pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23']
data.drop(drop_cols_1, axis=1, inplace=True)
```

```python
#Removendo as linhas com até 3 valores não nulos. Mínimo que julgo necessário para criar algum tipo de insight dada a base de dados estudada.
data.dropna(axis=0,how='all', thresh=3, inplace=True)
```

```python
#Somente com esta limpeza, diminuimos o tamanho da base de dados de 25722 linhas para 6301 linhas. 
len(data)
```

```python
#Rodando novamente o profile report, agora sem colunas e linhas que não fazem sentido
pp.ProfileReport(data)
```

```python
#Padronização dos dados (Activity)
print(set(data['Activity']))
```

```python
# Padronizando todos os textos que possuem as palavras chave 'fish, surf, dive, swin' para a cadetegoria 'fish'
data.loc[(data['Activity'].str.contains('fish')) & (data['Activity'].notnull()), 'Activity'] = 'Fishing'
data['Activity'] = data['Activity'].str.replace('Surf-skiing', 'Surf skiing')
data.loc[(data['Activity'].str.contains('Fish')) & (data['Activity'].notnull()), 'Activity'] = 'Fishing'
data.loc[(data['Activity'].str.contains('dive')) & (data['Activity'].notnull()), 'Activity'] = 'Diving'
data.loc[(data['Activity'].str.contains('Dive')) & (data['Activity'].notnull()), 'Activity'] = 'Diving'
data.loc[(data['Activity'].str.contains('diving')) & (data['Activity'].notnull()), 'Activity'] = 'Diving'
data.loc[(data['Activity'].str.contains('Diving')) & (data['Activity'].notnull()), 'Activity'] = 'Diving'
data.loc[(data['Activity'].str.contains('surf')) & (data['Activity'].notnull()), 'Activity'] = 'Surfing'
data.loc[(data['Activity'].str.contains('Surf')) & (data['Activity'].notnull()), 'Activity'] = 'Surfing'
data.loc[(data['Activity'].str.contains('swim')) & (data['Activity'].notnull()), 'Activity'] = 'Swimming'
data.loc[(data['Activity'].str.contains('Swim')) & (data['Activity'].notnull()), 'Activity'] = 'Swimming'
data.loc[(data['Activity'].str.contains('Swimming ')) & (data['Activity'].notnull()), 'Activity'] = 'Swimming'
data.loc[(data['Activity'].str.contains('Snorkeling')) & (data['Activity'].notnull()), 'Activity'] = 'Diving'
data.loc[(data['Activity'].str.contains('ward')) & (data['Activity'].notnull()), 'Activity'] = 'Warding'
data.loc[(data['Activity'].str.contains('Ward')) & (data['Activity'].notnull()), 'Activity'] = 'Warding'
data.loc[(data['Activity'].str.contains('boat')) & (data['Activity'].notnull()), 'Activity'] = 'Fishing'
data.loc[(data['Activity'].str.contains('Boat')) & (data['Activity'].notnull()), 'Activity'] = 'Fishing'
data.loc[(data['Activity'].str.contains('bath')) & (data['Activity'].notnull()), 'Activity'] = 'Bathing'
data.loc[(data['Activity'].str.contains('Bath')) & (data['Activity'].notnull()), 'Activity'] = 'Bathing'
data.loc[(data['Activity'].str.contains('ship')) & (data['Activity'].notnull()), 'Activity'] = 'Fishing'
data.loc[(data['Activity'].str.contains('Ship')) & (data['Activity'].notnull()), 'Activity'] = 'Fishing'
data.loc[(data['Activity'].str.contains('stand')) & (data['Activity'].notnull()), 'Activity'] = 'Standing'
data.loc[(data['Activity'].str.contains('Stand')) & (data['Activity'].notnull()), 'Activity'] = 'Standing'
data.loc[(data['Activity'].str.contains('Float')) & (data['Activity'].notnull()), 'Activity'] = 'Floating'
data.loc[(data['Activity'].str.contains('float')) & (data['Activity'].notnull()), 'Activity'] = 'Floating'
data.loc[(data['Activity'].str.contains('sail')) & (data['Activity'].notnull()), 'Activity'] = 'Sailing'
data.loc[(data['Activity'].str.contains('Sail')) & (data['Activity'].notnull()), 'Activity'] = 'Sailing'
```

```python
#Padronizando a valiavel Age
data.loc[(data['Age'].str.contains('months')) & (data['Age'].notnull()), 'Age'] = '0'
data.loc[(data['Age'].str.contains('Months')) & (data['Age'].notnull()), 'Age'] = '0'
data.loc[(data['Age'].str.contains(' 28')) & (data['Age'].notnull()), 'Age'] = '28'
data.loc[(data['Age'].str.contains('6½')) & (data['Age'].notnull()), 'Age'] = '6'
data.loc[(data['Age'].str.contains('60s')) & (data['Age'].notnull()), 'Age'] = '60'
```

```python
# Apos padronizar os dados, retirando colunas que não adicionam informações relevantes
cols_drop_2 = ['Type', 'Area', 'Location', 'Name', 'Injury', 'Investigator or Source']
data.drop(cols_drop_2, inplace= True, axis=1)
```
