## Projeto 5 - House Prices

* Diversas características de 1460 casas localizadas na cidade de Ames, estado de Iowa, nos Estados Unidos;
* Total de 81 variáveis, sendo muitas não numéricas;
* Pretende-se encontrar um modelo que preveja com boa precisão os preços de casas na região estudada.

## Primeiro Modelo: Principal component analysis (PCA)
#### Vantagens:
* Utilização de todas as variáveis disponíveis (80);
* Elimina o problema de multicolinearidade dos dados.

#### Desvantagens:
* Só é possível fazer previsões, perde-se o impacto de cada variável no preço das casas



* R2 no set de treino: 0.95
* R2 no set de teste: 0.66


## Segundo Modelo: Regressão Linear com variáveis fortemente correlacionadas com 'Sales Price'
#### Vantagens:
* Podemos observar o peso de cada variável nos preços.

#### Desvantagens:
* Nem todas as variáveis são consideradas no modelo;
* Pode existir multicolinearidade entre os dados;
* Somente foram utilizadas variáveis numéricas.



* R2 no set de treino: 0.80
* R2 no set de teste: 0.72

## Terceiro modelo: Regressão Linear com variáveis selecionadas

#### Vantagens:
* Podemos analisar os pesos de variáveis que acreditamos serem relevantes nos preços das casas;
* A seleção permite nos afastarmos do problema de multicolinearidade.

#### Desvantagens:
* Omissão de muitas variáveis.



* R2 no set de treino: 0.69
* R2 no set de teste: 0.72
